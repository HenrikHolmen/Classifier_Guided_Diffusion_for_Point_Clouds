### Loading data and plotting

```python
import numpy as np

# Load the numpy file
data = np.load('out_gen_num200.npy')

# Display the shape of the loaded data to understand its structure
data.shape
```

### Plotting the first 3 frames

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 

# Function to plot a single frame
def plot_frame(data, frame_number):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(data[frame_number, :, 0], data[frame_number, :, 1], data[frame_number, :, 2], s=1)
    ax.set_title(f'Frame {frame_number}')
    plt.show()

# Plot the first 3 frames
for i in range(3):
    plot_frame(data, i)
```
