### Plotting Coverage over iterations

```python
import matplotlib.pyplot as plt

# File path
file_path = 'Converage_150_1040k.txt'

# Read the file content
with open(file_path, 'r', encoding="utf-8") as f:
    lines = f.readlines()

# Extract CD values from each line
cd_values = [float(line.strip().split()[-1]) for line in lines]

# Generate iterations (starting from 10000, 20000, ..., etc.)
iterations = [(i + 1) * 10000 for i in range(len(cd_values))]

# Plotting the validation curve for Chamfer Distance (CD)
plt.figure(figsize=(10, 5))
plt.plot(iterations, cd_values, label='Chamfer Distance (CD)', marker='o')
plt.xlabel('Iterations')
plt.ylabel('Chamfer Distance (CD)')
plt.title('Validation Chamfer Distance Curve')
plt.legend()
plt.grid(True)
plt.savefig('Validation_CD_curve.png', dpi=500)
plt.show()
```

```python
# Plotting the validation curve for Chamfer Distance (CD) with log scale on x-axis
plt.figure(figsize=(10, 5))
plt.plot(iterations, cd_values, label='Chamfer Distance (CD)', marker='o')
plt.xlabel('Iterations (log scale)')
plt.ylabel('Chamfer Distance (CD)')
plt.xscale('log')
plt.title('Validation Chamfer Distance Curve (Log Scale)')
plt.legend()
plt.grid(True)
plt.savefig('Validation_CD_curve_log.png', dpi=500)
plt.show()
```

### Combined Validation Curves normal scale

```python
import matplotlib.pyplot as plt

# File paths
file_paths = {
    'Coverage': 'Converage_150_1040k.txt',
    'MinMatDis': 'MinMatDis_150_1040k.txt',
    '1NN-Accur': '1NN_150_1040k.txt',
    'Jensen-Shannon': 'JSD_150_1040k.txt'
}

# Read the file contents
data = {}
for key, file_path in file_paths.items():
    with open(file_path, 'r', encoding="utf-8") as f:
        lines = f.readlines()
        values = [float(line.strip().split()[-1]) for line in lines]
        data[key] = values

# Generate iterations (starting from 10000, 20000, ..., etc.)
iterations = [(i + 1) * 10000 for i in range(len(data['Coverage']))]

# Plotting the combined figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Coverage Chamfer Distance (Normal Scale)
axs[0, 0].plot(iterations, data['Coverage'], label='Coverage Chamfer Distance (CD)', marker='o')
axs[0, 0].set_xlabel('Iterations')
axs[0, 0].set_ylabel('Chamfer Distance (CD)')
axs[0, 0].set_title('Coverage Chamfer Distance')
axs[0, 0].legend()
axs[0, 0].grid(True)

# MinMatDis Chamfer Distance (Normal Scale)
axs[0, 1].plot(iterations, data['MinMatDis'], label='MinMatDis Chamfer Distance (CD)', marker='o')
axs[0, 1].set_xlabel('Iterations')
axs[0, 1].set_ylabel('Chamfer Distance (CD)')
axs[0, 1].set_title('MinMatDis Chamfer Distance')
axs[0, 1].legend()
axs[0, 1].grid(True)

# 1NN-Accur Chamfer Distance (Normal Scale)
axs[1, 0].plot(iterations, data['1NN-Accur'], label='1NN-Accur Chamfer Distance (CD)', marker='o')
axs[1, 0].set_xlabel('Iterations')
axs[1, 0].set_ylabel('Chamfer Distance (CD)')
axs[1, 0].set_title('1NN-Accur Chamfer Distance')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Jensen-Shannon Divergence (Normal Scale)
axs[1, 1].plot(iterations, data['Jensen-Shannon'], label='Jensen-Shannon Divergence', marker='o')
axs[1, 1].set_xlabel('Iterations')
axs[1, 1].set_ylabel('Jensen-Shannon Divergence')
axs[1, 1].set_title('Jensen-Shannon Divergence')
axs[1, 1].legend()
axs[1, 1].grid(True)

plt.tight_layout()
plt.savefig('Combined_Validation_Curves.png', dpi=500)
plt.show()
```
